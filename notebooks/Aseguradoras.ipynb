```python
import pandas as pd
```

```python

url = "https://raw.githubusercontent.com/FernandoCasco/etl-data-pipeline/refs/heads/main/data/raw/aseguradoras.csv"

aseguradoras = pd.read_csv(url)

aseguradoras.head()
```

```python
# Exploracion de datos

aseguradoras.info()
aseguradoras.columns
aseguradoras.isnull().sum()
```

```python

#Limpieza de datos

def limpiar_dataframe(df):

    df.columns = df.columns.str.strip().str.lower()

    for col in df.select_dtypes(include='object').columns:
        df[col] = df[col].astype(str).str.strip()

    df = df.replace(r'^\s*$', pd.NA, regex=True)

    df = df.drop_duplicates()

    return df
```

```python
aseguradoras = limpiar_dataframe(aseguradoras)
```

```python
aseguradoras.info()

aseguradoras.isnull().sum()

aseguradoras.head()
```

```python
aseguradoras['pais'] = aseguradoras['pais'].replace({
    "ElSalvador": "El Salvador"
})
```

```python
#Transformaciones Normalizar nombres de países

aseguradoras['pais'] = aseguradoras['pais'].replace({
    "ElSalvador": "El Salvador"
})
```

```python
#Normalizar texto (capitalización)
aseguradoras['pais'] = aseguradoras['pais'].str.title()
aseguradoras['nombre'] = aseguradoras['nombre'].str.title()
```

```python
#Estandarizar valores de rating

aseguradoras['rating_riesgo'] = aseguradoras['rating_riesgo'].str.capitalize()
```

```python
#Validar tipos de datos

aseguradoras['id_aseguradora'] = aseguradoras['id_aseguradora'].astype(int)
```

```python
#Exportacion de de archivos curated

aseguradoras.to_csv("aseguradoras_curated.csv", index=False)
```
